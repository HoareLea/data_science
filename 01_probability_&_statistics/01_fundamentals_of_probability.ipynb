### **Data, Sample Space & Events**
**Data** are organised information relating to a thing or process. The data can be **continuous** meaning they can take any value (possibly within a range) or **discrete** meaning they can only take a finite number of possible values. For example height is a continuous variable since it can be any positive value whereas traffic light colour is discrete because it can only take three values (red, amber, green). Note data is the plural form of the singular datum.

### **Probability Terminology & Notation**
We denote the probability of an event $A$ occuring as $\mathbb{P}(A)$

We denote the probability of an event $A$ not occuring as $\mathbb{P}(A')$ or $\mathbb{P}(A^c)$ where the $c$ stands for 'compliment'

The **intersection** of two events $A,B$ is the probability of both $A$ and $B$ occuring and is written $\mathbb{P}(A \cap B)$

The **union** of two events $A,B$ is the probability of either $A$ or $B$ occuring and is written $\mathbb{P}(A \cup B)$

The probability of event $A$ occuring given event $B$ has occured is written $\mathbb{P}(A|B)$

Two event $A,B$ are said to be **mutually exclusive** if $\mathbb{P}(A \cap B) = 0$

### **Axioms of Probability**
The axioms of probability are the basic laws we assume when working with probability and statistics. 

Let $S$ be the sample space and $A,B$ be events. Then the laws are:
1. The probability of an event is non-zero
2. The probability of all possible events is 1
3. For mutually exclusive events, the proability of them both occuring is the sum of their probabilities

Mathemtically, these axioms can be written:
1. $\mathbb{P}(A)>=0$
2. $\mathbb{P}(S)=1$
3. $\mathbb{P}(A \cup B) = \mathbb{P}(A) + \mathbb{P}(B)$ for $A,B$ mutually exclusive

### **Corollaries of the Axioms of Probability**

$\mathbb{P}(A^c) = 1- \mathbb{P}(A)$

$\mathbb{P}(A \cup B) = \mathbb{P}(A) + \mathbb{P}(B) - \mathbb{P}(A \cap B)$ (prevents double counting intersection)

$\mathbb{P}(A|B) = \frac{\mathbb{P}(A \cap B)}{\mathbb{P}(B)}$ (for $\mathbb{P}(B) > 0$)

$\mathbb{P}(A \cap B) = \mathbb{P}(A|B)\mathbb{P}(B)$

### **Independence & Dependence**
**Independence** is a central idea within staistics. Intuitively, it captures the notion that the outcome of one event does not effect the outcome for another event. For instance, when flipping a fair coin, the probability of heads is unrelated to whether or not we got heads last on the previous flip. The converse of independence is dependence and relates to events whose outcomes do affect one another. For example the temperature in London today is related to the temperature yesterday or on this day last year. 

Mathematically we capture the idea of independence as follows:

Two events, $A,B$ are independent if and only if $\mathbb{P}(A\cap B) = \mathbb{P}(A)\mathbb{P}(B)$

To see why this captures the notion of the two events being unrelated notice that for independent event $A,B$ we have:

$\mathbb{P}(A|B) = \frac{\mathbb{P}(A \cap B)}{\mathbb{P}(B)} = \frac{\mathbb{P}(A)\mathbb{P}(B)}{\mathbb{P}(B)} = \mathbb{P}(A)$

Hence the probability of event $A$ given that $B$ has occured is just the probability of event $A$; the chance that $A$ occurs is unrelated to whether or not $B$ has occured. 

### **Law of Total Probability**
The motivation for the Law of Total Probability is suppose you want to compute the probability of some event $A$ but calcualting $\mathbb{P}(A)$ directly is difficult. However, you know that the world can be divided into distinct â€œcasesâ€ $B_1, \cdots , B_n$ such that they are:

Instead of trying to compute $\mathbb{P}(A)$ we instead compute $\mathbb{P}(A|B_i)$ for each $B_i$, weighted by how likely each case is. 

**Statement**
Suppose $B_1, \cdots , B_n$ are events which partition a sample space $S$ then:

$\mathbb{P}(A) = \sum_{i=1}^{n} \mathbb{P}(A \mid B_i)\, \mathbb{P}(B_i)$

**Proof**

Let $B_1, \cdots , B_n$ be a partition of $S$ with $\mathbb{P}(B_i) > 0$ for $i=1, \cdots ,n$

$A = A \cap S = A \cap \left(\bigcup_{i=1}^n B_i\right) = \bigcup_{i=1}^n (A \cap B_i)$

Since the $B_i$ are disjoint (i.e. mutually exclusive), the events $A \cap B_i$ are disjoint 

$\mathbb{P}(A)=\mathbb{P}\!\left(\bigcup_{i=1}^n (A \cap B_i)\right)=\sum_{i=1}^n \mathbb{P}(A \cap B_i)$

$\mathbb{P}(A \mid B_i)=\frac{\mathbb{P}(A \cap B_i)}{\mathbb{P}(B_i)}\quad \Rightarrow \quad\mathbb{P}(A \cap B_i)=\mathbb{P}(A \mid B_i)\,\mathbb{P}(B_i)$

$\mathbb{P}(A)=\sum_{i=1}^n\mathbb{P}(A \mid B_i)\,\mathbb{P}(B_i)$
