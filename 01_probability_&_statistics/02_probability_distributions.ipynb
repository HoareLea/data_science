### **Experiments & Random Variables**
An **experiment** is a repeatable process with an uncertain outcome but a well-defined set of possible outcomes (the sample space $S$). 

A **random variable** is a function $X: S \rightarrow \mathbb{R}$ that assigns a numerical value to each outcome of that experiment, allowing us to analyse the randomness mathematically. 

In other words, the experiment generates the uncertainty, and the random variable translates that uncertainty into numbers we can study.

### **Probability Distributions**
A random variable $X$ is said to follow a probability distribution which is a function $\mathbb{P}(X=x): \mathbb{R} \rightarrow [0,1]$ that describes how probability is assigned to it's possible values. 

For example when we flip a unfair coin the sample space is $S=\{Heads, Tails\}$ and the random variable X could map these values to $\{1, 0\}$. 

Thus the probability distribution is a function $\mathbb{P}(X=x): \{1, 0\} \rightarrow [0,1]$. 

In this example suppose the probability of heads was 0.7 and tails 0.3. We could write the probability distribution as follows:

$\mathbb{P}(X=x) = \begin{cases} 0.7 & \text{if } x = 1 \\ 0.3 & \text{otherwise} \end{cases}$


### **PMF, PDF & CDF**
If $X$ is discrete then $p_X(x) = \mathbb{P}(X = x)$ is called the **probability mass function (PMF)**. 
Probabilities are directly assigned to specific values, so $\mathbb{P}(X = x)$ can be positive. 
The PMF satisfies $\sum_x p_X(x) = 1$.

If $X$ is continuous then we define the **probability density function (PDF)**, denoted $f_X(x)$. 
Probabilities are assigned to intervals rather than individual points, so $\mathbb{P}(X = x) = 0$ for all $x$. 
Probabilities are computed via integration $\mathbb{P}(a \le X \le b) = \int_a^b f_X(x),dx$ and the PDF satisfies $\int_{-\infty}^{\infty} f_X(x),dx = 1$.

The **cumulative distribution function (CDF)** is defined as $F_X(x) = \mathbb{P}(X \le x)$

We can model random variables as following a particular probability distribution which we can define using one of these functions. 

### **Expectation (Mean)**
Intuitively, we understand the **expected value** of a distribution as the average or centre of mass of the data.

The expected value of a random variable $X$ is define as follows:
- For discrete $X$:  $E[X] = \sum_{x} x P(X = x)$
- For continuous $X$: $E[X] = \int_{-\infty}^{\infty} x f(x), dx$

From this definition, we can derive the **linearity of expectation**:
- $E[aX + b] = aE[X] + b$
- $E[X + Y] = E[X] + E[Y]$
- $E\left[\sum_{i=1}^{n} X_i\right] = \sum_{i=1}^{n} E[X_i]$

### **Variance**
Intuitively, we understand the **variance** to represent how spread out the data is. The definition captures this intuition: we look at the average squared distance from the mean. Note that we square it to make it positive since we aren't interested in whether the data is above or below the mean, just how spread out from the mean it is. 

The variance of a random variable $X$ is define as follows:
$Var(X)=E[(Xâˆ’\mu)^2]

We can also express the variance as follows:
$Var(X)=E[X^2] - (E[X])^2$

This formulation makes computation easier and can be derived using the linearity of expectation: 
$E[X^2 - 2\mu X + \mu^2]$ = $E[X^2] - 2\mu E[X] + E[\mu^2]$ = $E[X^2] - 2\mu^2 + \mu^2$ = $E[X^2] - \mu^2$ = $E[X^2] - (E[X])^2$

Similarly we can derive:
$Var(aX+b) = a^2Var(X)$

This shows that translating the data ($+b$) doesn't affect the variance which intuitively makes sense; moving teh data doesn't affect how spread out from it's mean it is. However scaling the data $aX$ does affect that. 

We often work with the **standard deviation** which is defined as $\sigma = \sqrt{Var(X)}$

The units of standard deviation are on the same scale as X (unlike with Var) so this allows us to compare the measure of spread across different variables (e.g. *"the spread of heights is greater than the spread of weights"*)

### **Covariance**
Covariance measures how two random variables move together.
- If large values of $X$ tend to occur with large values of $Y$, covariance is positive.
- If large values of $X$ tend to occur with small values of $Y$, covariance is negative.
- If there is no systematic linear co-movement, covariance is close to zero.
It captures linear association only.

The covariance between two random variables $X$ and $Y$ is defined as:

$\mathrm{Cov}(X,Y) = E\left[(X - E[X])(Y - E[Y])\right]$

This measures the expected product of their deviations from their means.

It can also be expressed as:
$\mathrm{Cov}(X,Y) = E[XY] - E[X]E[Y]$

Some key properties of covariance are as follows:
- $\mathrm{Cov}(X,Y) = \mathrm{Cov}(Y,X)$
- $\mathrm{Var}(X) = \mathrm{Cov}(X,X)$
- $\mathrm{Cov}(aX + b, Y) = a \mathrm{Cov}(X,Y)$
- $\mathrm{Cov}(X,Y) = 0$
- $\mathrm{Var}(X + Y) = \mathrm{Var}(X) + \mathrm{Var}(Y) + 2 \mathrm{Cov}(X,Y)$

### **Correlation**
The issue with covariance is that it is scale dependent. This means the covariance of weight and heigh measured in cm will be different to the covar weight and heigh measured in m. 

The correlation between two random variables $X$ and $Y$ is defined as:
$\rho_{X,Y} = \dfrac{\mathrm{Cov}(X,Y)}{\sqrt{\mathrm{Var}(X)} \sqrt{\mathrm{Var}(Y)}}$

This measures covariance on a standardised scale, removing the effect of units.

It can also be expressed as:

$\rho_{X,Y} = \dfrac{E[XY] - E[X]E[Y]}{\sqrt{\mathrm{Var}(X)} \sqrt{\mathrm{Var}(Y)}}$

Some key properties of correlation are as follows:
- $-1 \le \rho_{X,Y} \le 1$
- $\rho_{X,Y} = \rho_{Y,X}$
- $\rho_{X,X} = 1$
- $\rho_{X,Y} = 0$ does not necessarily imply independence
- $\rho_{aX + b, \, cY + d} = \begin{cases} \rho_{X,Y} & \text{if } ac > 0 \\ -\rho_{X,Y} & \text{if } ac < 0 \end{cases}$


### **Bernoulli Distribution**
A Bernoulli random variable (discrete) models a single binary outcome with probability of sucess $p$. Note typically we model sucess as $X=1$ and failure as $X=0$.

The PMF is given by:
$P(X=x) = p^x(1-p)^{1-x}, \quad 0 \le p \le 1$

Mean: $E[X] = p$
Variance: $\mathrm{Var}(X) = p(1-p)$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli

def plot_bernoulli(ps=(0.2, 0.5, 0.8)):
    x = np.array([0, 1])
    plt.figure()
    for p in ps:
        pmf = bernoulli.pmf(x, p)
        plt.plot(x, pmf, marker='o', linestyle='None', label=f"p={p}")
    plt.xticks([0, 1])
    plt.xlabel("x")
    plt.ylabel("PMF: P(X=x)")
    plt.title("Bernoulli(p)")
    plt.legend()
    plt.tight_layout()
    plt.show()

plot_bernoulli()
```

### **Binomial Distribution**
A Binomial random variable (discrete) models the number of successes in $n$ independent Bernoulli trials, each with probability of sucess $p$.

The PMF is given by:
$P(X=k) = \binom{n}{k} p^k (1-p)^{n-k}, \quad 0 \le p \le 1$

Mean: $E[X] = np$
Variance: $\mathrm{Var}(X) = np(1-p)$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

def plot_binomial(n=20, ps=(0.2, 0.5, 0.8)):
    k = np.arange(0, n + 1)
    plt.figure()
    for p in ps:
        pmf = binom.pmf(k, n, p)
        plt.plot(k, pmf, marker='o', linestyle='-', label=f"n={n}, p={p}")
    plt.xlabel("k")
    plt.ylabel("PMF: P(X=k)")
    plt.title("Binomial(n, p)")
    plt.legend()
    plt.tight_layout()
    plt.show()

plot_binomial()
```

### **Poisson Distribution**
A Poisson random variable (discrete) models the count of an event in a fixed interval where the events occur independently at an average rate of $\lambda$.

The PMF is given by:
$P(X=k) = \frac{\lambda^k e^{-\lambda}}{k!}, \quad \lambda > 0$

Mean: $E[X] = \lambda$
Variance: $\mathrm{Var}(X) = \lambda$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

def plot_poisson(lambdas=(1, 4, 10)):
    k_max = int(max(lambdas) + 4*np.sqrt(max(lambdas)))
    k = np.arange(0, k_max + 1)

    plt.figure()
    for lam in lambdas:
        pmf = poisson.pmf(k, lam)
        plt.plot(k, pmf, marker='o', linestyle='-', label=f"Î»={lam}")
    plt.xlabel("k")
    plt.ylabel("PMF: P(X=k)")
    plt.title("Poisson(Î»)")
    plt.legend()
    plt.tight_layout()
    plt.show()

plot_poisson()
```

### **Uniform Distribution**
A Uniform random variable (continuous) models where every value in an interval $[a,b]$ is equally liekly.

The PDF is given by:
$f(x) = \frac{1}{b-a}$ for $a \le x \le b$, and $0$ otherwise.

Mean: $E[X] = \frac{a+b}{2}$
Variance: $\mathrm{Var}(X) = \frac{(b-a)^2}{12}$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

def plot_continuous_uniform(params=((0, 1), (-1, 1), (2, 5))):
    plt.figure()
    for a, b in params:
        x = np.linspace(a - 0.5*(b-a), b + 0.5*(b-a), 500)
        pdf = uniform.pdf(x, loc=a, scale=(b-a))
        plt.plot(x, pdf, label=f"a={a}, b={b}")
    plt.xlabel("x")
    plt.ylabel("PDF: f(x)")
    plt.title("Uniform(a, b) continuous")
    plt.legend()
    plt.tight_layout()
    plt.show()

plot_continuous_uniform()
```

### **Normal (Gaussian) Distribution**
A Normal random variable (continuous) models the classic bell-shaped variation we frequently observe in many processes.

The PDF is given by:
$f(x) = \frac{1}{\sigma \sqrt{2\pi}} \exp\left(-\frac{(x-\mu)^2}{2\sigma^2}\right), \quad \sigma > 0$

Mean: $E[X] = \mu$
Variance: $\mathrm{Var}(X) = \sigma^2$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_normal(cases=((0, 1), (0, 2), (2, 1))):
    # choose x-range to cover all cases
    mus = [mu for mu, s in cases]
    sigs = [s for mu, s in cases]
    lo = min(mus) - 4*max(sigs)
    hi = max(mus) + 4*max(sigs)
    x = np.linspace(lo, hi, 800)

    plt.figure()
    for mu, s in cases:
        pdf = norm.pdf(x, loc=mu, scale=s)
        plt.plot(x, pdf, label=f"Î¼={mu}, Ïƒ={s}")
    plt.xlabel("x")
    plt.ylabel("PDF: f(x)")
    plt.title("Normal(Î¼, Ïƒ)")
    plt.legend()
    plt.tight_layout()
    plt.show()

plot_normal()
```

### **Exponential Distribution**
An Exponential random variable (continuous) models the waiting time until the next event in a Poisson process.

The PDF is given by:
$f(x) = \lambda e^{-\lambda x}, \quad x \ge 0,\ \lambda > 0$

Mean: $E[X] = \frac{1}{\lambda}$
Variance: $\mathrm{Var}(X) = \frac{1}{\lambda^2}$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

def plot_exponential(lambdas=(0.5, 1.0, 2.0)):
    x = np.linspace(0, 10, 800)
    plt.figure()
    for lam in lambdas:
        # scipy expon uses scale = 1/lambda
        pdf = expon.pdf(x, scale=1/lam)
        plt.plot(x, pdf, label=f"Î»={lam}")
    plt.xlabel("x")
    plt.ylabel("PDF: f(x)")
    plt.title("Exponential(Î»)")
    plt.legend()
    plt.tight_layout()
    plt.show()

plot_exponential()
```

### **Gamma Distribution**
A Gamma random variable (continuous) models the waiting time until the $k^{th}$ event in a Poisson process.

The PDF is given by:
$f(x) = \frac{1}{\Gamma(k),\theta^k} x^{k-1} e^{-x/\theta}, \quad x \ge 0,\ k>0,\ \theta>0$

Mean: $E[X] = k\theta$
Variance: $\mathrm{Var}(X) = k\theta^2$
