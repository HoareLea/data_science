### **Continuous Integration**
In software engineering there are several key principles and best practices which help to ensure software products are developed in a way which promotes quality. One key idea is that proposed changes to code are validated to ensure they do not break anything. This concept is known as **Continuous Integration (CI)**. In practice this means that anytime someone opens a PR to propose changes, the system automatically:
- Installs dependencies (to check they work)
- Runs tests
- Lints code (to check for errors)
- Checks artifacts build (i.e. Docker images)

The goal of CI is to catch problems early, before they cause serious problems. Developers are also encouraged to commit small changes often, rather than large changes infrequently. This has a number of benefits:
- It forces developers to break more complex changes down into smaller testable parts
- It tests the changes often so erros are identified early 
- It aids debugging; it can be very difficult to identify where they bug is in a large and complex change

Overall CI makes development faster, easier and of higher quality
